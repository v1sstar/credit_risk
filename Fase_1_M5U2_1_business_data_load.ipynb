**¿Qué solución basada en Deep Learning queremos implementar?**

La solución consiste en un modelo predictivo de clasificación binaria de riesgo crediticio basado en **Redes Neuronales Profundas (Deep Neural Networks - DNNs)**, que analizará el perfil de un solicitante de crédito para determinar la probabilidad de que incurra en impago (default).

El modelo clasificará a cada solicitante en una de dos categorías: "alto riesgo" (predicción de impago) o "bajo riesgo" (predicción de pago oportuno). Se opta por una arquitectura de red neuronal profunda debido a su capacidad para superar varias limitaciones inherentes a los modelos estadísticos tradicionales que se han estado implementando comúnmente en la industria.

Los modelos estadísticos clásicos, como la regresión logística, asumen una relación lineal entre las variables predictoras y el resultado. Sin embargo, el riesgo crediticio es un fenómeno que puede tener un comportamiento no lineal, por lo que las redes neuronales profundas, gracias a sus múltiples capas y funciones de activación no lineales, pueden aprender y modelar estas relaciones complejas y no lineales de forma automática, sin necesidad de que un analista las especifique. Además, las redes neuronales profundas descubren interacciones entre variables de manera implícita a través de sus capas ocultas. Cada capa aprende representaciones de los datos cada vez más abstractas y complejas, capturando patrones sutiles que serían difíciles de identificar y codificar manualmente. Finalmente, al tratarse de modelos no paramétricos, no se requieren suposiciones estrictas sobre la distribución de los datos, lo que los hace más flexibles y robustos, permitiendo una mejor capacidad de generalización.

**¿Qué se hará?**

Se llevará a cabo el ciclo completo de un proyecto de ciencia de datos con enfoque de la metodología CRISP-DM, abarcando las siguientes fases:

1. Análisis exploratorio y Preprocesamiento de los datos

Se realizará un análisis del conjunto de datos Credit Risk Dataset para entender sus características, distribuciones, y la presencia de datos atípicos o faltantes. Se aplicarán técnicas de limpieza, imputación, y en caso de ser necesario, codificación de variables categóricas (ej. One-Hot Encoding), o escalado de características numéricas, con el objetivo de preparar los datos para el entrenamiento.

2. Diseño e implementación del modelo

Se explorarán y diseñarán diferentes arquitecturas de redes neuronales profundas, experimentando con el número de capas, la cantidad de neuronas por capa, las funciones de activación y las técnicas de regularización (como Dropout) para evitar el sobreajuste.

El conjunto de datos se dividirá en subconjuntos de entrenamiento, validación y prueba para garantizar una evaluación objetiva.

Se entrenarán los modelos candidatos y se optimizarán sus hiperparámetros utilizando el conjunto de validación.

3. Modelación y evaluación del modelo final

El rendimiento del modelo seleccionado se medirá sobre el conjunto de prueba. Se utilizarán métricas de evaluación clave para problemas de clasificación desbalanceada, como la matriz de confusión, Precisión, Recall, F1-Score y el Área Bajo la Curva ROC (AUC-ROC), para obtener una visión de su capacidad predictiva.

- Evaluación automatizada de solicitudes: Al recibir una nueva solicitud de crédito, el sistema podrá procesar los datos del cliente y entregar una puntuación de riesgo (probabilidad de impago) junto con una recomendación (ej. "Aprobar", "Rechazar", "Revisión Manual").

Por lo que, el producto final será una herramienta analítica avanzada que potenciará la toma de decisiones en el sector bancario y financiero, permitiendo una gestión del riesgo crediticio más precisa, eficiente e incluso rentable.
